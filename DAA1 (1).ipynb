##### problem Statement
Write a program to find Minimum and Maximum element in the given array using
Min-Max Algorithm based on Divide and Conquer Strategy.

```python
import random
import time


# In[5]:


def linearMethod(array,n):
    small=big=array[0]
    for i in range(n):
        if small>array[i]:
            small=array[i]
            i+=1

        elif big<array[i]:

            big=array[i]
            i+=1
    print("******************************************MINIMUM ELEMENT BY USING LINEAR METHOD IS *********************************************************")
    print(small)
    print("**********************************MAXIMUM ELEMENT BY USING LINEAR METHOD IS*******************************************************************************")
    print(big)

def printList(array):
    for i in range(len(array)):
        print(array[i],end=" ")
    print()

if __name__ == '__main__':

    for _ in range(int(input("enter the test  case:"))):

        n=int(input("enter the size:"))
        array=random.sample(range(1,n+1),n)

        print("the original array is",array)
        print("*****************************")
        start_time=time.perf_counter()
        linearMethod(array,n)



        end_time=time.perf_counter()
        print("the taken in linear method is ",end_time-start_time)
        print("*****************************")
        print(printList(array))
```

```python
## import random
import time


# In[5]:


def MergeSort(array):
    if len(array)>1:
        mid=len(array)//2
        h1=array[:mid]
        h2=array[mid:]


        MergeSort(h1)
        MergeSort(h2)

        p=q=r=0

        while p<len(h1) and q<len(h2):
            if h1[p]<h2[q]:
                array[r]=h1[p]
                p+=1
            else:
                array[r]=h2[q]
                q+=1
            r+=1

        while p<len(h1):
            array[r]=h1[p]
            p+=1
            r+=1
        while q<len(h2):
            array[r]=h2[q]
            q+=1
            r+=1

def printList(array):
    for i in range(len(array)):
        print(array[i],end=" ")
    print()

if __name__ == '__main__':

    for _ in range(int(input("enter the test  case:"))):

        n=int(input("enter the size:"))
        array=random.sample(range(1,n+1),n)

        print("the original array is",array)
        print("*****************************")
        start_time=time.perf_counter()
        MergeSort(array)
        print("sorted array",array)
        print("*****************************")
        print("Minimum element is ",array[0])
        print("Maximum element is ",array[n-1])



        end_time=time.perf_counter()
        print("the taken in merge sort is ",end_time-start_time)
        print("*****************************")
        print(printList(array))
```
